# Splitting the Genre of Category 

```python
import re
```

```python
def split_Camel_Case(text):
    return re.findall(r'[A-Z][a-z]*', text)
```

```python
separated_list = [split_Camel_Case(item) for item in df["Genre"]]
```

```python
separated_list
```

```python
df["Genre"] = separated_list
```

```python
df
```

```python
df.info()
```

```python
df.drop("index", axis=1,  inplace=True)
```

```python
# cleaning User-rating column
l = []
for i in df['User-Rating']:
    l.append(i.replace("/10",""))
```

```python
l = [float(i) for i in l]
```

```python
l
```

```python
df['User-Rating'] = l
```

```python
# types of Genres 
genres_list = []
for i in df['Genre']:
    for j in i:
        genres_list.append(j);
genres_list = set(genres_list)
```
